#### Functions to loop through all social welfare sites

```python
# modified to include distance from site and address
def getNearbyVenues(names, subdist, latitudes, longitudes, radius=500):

#    SiteName	Province	District	SubDist	Xlong	YLat
    
    venues_list=[]
    for name, sd, lat, lng  in zip(names, subdist, latitudes, longitudes):
        print(name, sd)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            sd,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['id'],
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['location']['distance'], 
            v['venue']['location']['formattedAddress'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['SiteName', 
                  'SubDistrict', 
                  'Site Latitude', 
                  'Site Longitude', 
                  'Venue', 
                  'Venue_id', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Distance',
                  'Venue Address',
                  'Venue Category']
    
    return(nearby_venues)
```

### Convert into a pandas Dataframe

```python
# Function to sort venues in decending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# get dbs of top 10 venues per social welfare sites

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['SiteName']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
sites_venues_sorted = pd.DataFrame(columns=columns)
sites_venues_sorted['SiteName'] = coj_grouped['SiteName']

for ind in np.arange(coj_grouped.shape[0]):
    sites_venues_sorted.iloc[ind, 1:] = return_most_common_venues(coj_grouped.iloc[ind, :], num_top_venues)

sites_venues_sorted.head()
```
